```python
# Import library
import statsmodels.formula.api as smf

# Simple Linear Regression
model = smf.ols('output ~ input', data = df).fit()
model.summary()

pred1 = model.predict(pd.DataFrame(df['input']))
```

```python
# Regression Line
plt.scatter(df.input, df.output)
plt.plot(df.input, pred1, "r")
plt.legend(['Predicted line', 'Observed data'])
plt.show()

# Error calculation
res1 = df.output - pred1
res_sqr1 = res1 * res1
mse1 = np.mean(res_sqr1)
rmse1 = np.sqrt(mse1)
rmse1
```

```python
plt.scatter(x = np.log(df['output']), y = df['input'], color = 'brown')
np.corrcoef(np.log(df.output), df.input) #correlation

model2 = smf.ols('output ~ np.log(input)', data = df).fit()
model2.summary()

pred2 = model2.predict(pd.DataFrame(df['input']))
```

```python
# Regression Line
plt.scatter(np.log(df.input), df.output)
plt.plot(np.log(df.input), pred2, "r")
plt.legend(['Predicted line', 'Observed data'])
plt.show()

# Error calculation
res2 = df.output - pred2
res_sqr2 = res2 * res2
mse2 = np.mean(res_sqr2)
rmse2 = np.sqrt(mse2)
rmse2
```

```python
plt.scatter(x = df['input'], y = np.log(df['output']), color = 'orange')
np.corrcoef(df.input, np.log(df.output)) #correlation

model3 = smf.ols('np.log(output) ~ input', data = df).fit()
model3.summary()

```

```python
pred3 = model3.predict(pd.DataFrame(df['input']))
pred3_at = np.exp(pred3)
pred3_at

# Regression Line
plt.scatter(df.input, np.log(df.output))
plt.plot(df.output, pred3, "r")
plt.legend(['Predicted line', 'Observed data'])
plt.show()

# Error calculation
res3 = df.output - pred3_at
res_sqr3 = res3 * res3
mse3 = np.mean(res_sqr3)
rmse3 = np.sqrt(mse3)
rmse3
```

```python
#### Polynomial transformation
# x = waist; x^2 = waist*waist; y = log(at)
model4 = smf.ols('np.log(output) ~ input + I(input*input)', data = df).fit()
model4.summary()

pred4 = model4.predict(pd.DataFrame(df))
pred4_at = np.exp(pred4)
pred4_at


# Regression line
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X = df.iloc[:, 0:1].values
X_poly = poly_reg.fit_transform(X)
# y = wcat.iloc[:, 1].values


plt.scatter(df.input, np.log(df.output))
plt.plot(X, pred4, color = 'red')
plt.legend(['Predicted line', 'Observed data'])
plt.show()



```

```python
# Error calculation
res4 = df.output - pred4_at
res_sqr4 = res4 * res4
mse4 = np.mean(res_sqr4)
rmse4 = np.sqrt(mse4)
rmse4
```

```python
# Choose the best model using RMSE
data = {"MODEL":pd.Series(["SLR", "Log model", "Exp model", "Poly model"]), "RMSE":pd.Series([rmse1, rmse2, rmse3, rmse4])}
table_rmse = pd.DataFrame(data)
table_rmse
```

```python
###################
# The best model

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size = 0.2)

finalmodel = smf.ols('output ~ input', data = train).fit()
finalmodel.summary()

```

```python
# Predict on test data
test_pred = finalmodel.predict(pd.DataFrame(test))
pred_test_output = np.exp(test_pred)
pred_test_output

# Model Evaluation on Test data
test_res = test.output - pred_test_output
test_sqrs = test_res * test_res
test_mse = np.mean(test_sqrs)
test_rmse = np.sqrt(test_mse)
test_rmse




```

```python
# Prediction on train data
train_pred = finalmodel.predict(pd.DataFrame(train))
pred_train_output = np.exp(train_pred)
pred_train_output

# Model Evaluation on train data
train_res = train.output - pred_train_output
train_sqrs = train_res * train_res
train_mse = np.mean(train_sqrs)
train_rmse = np.sqrt(train_mse)
train_rmse
```
